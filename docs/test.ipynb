```python
import matplotlib
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
%matplotlib inline

font_location = '/Users/seongyeon/Library/Fonts/경기천년제목OTF_Light.otf'
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

# Create a positioning graph for ease of comparison.
fig, ax = plt.subplots()

# Labels for competitors and our app
labels = ['StoryCorps', 'Healium', '아름다운 이야기 할머니', 'Memoir Writing Services', 
          'AI Story Writer', '오이', '시놀', '포페런츠', '우리의 앱']

# X-axis: Technology Integration (Low -> High)
x = [2, 9, 1, 3, 7, 4, 5, 3, 8]

# Y-axis: Content Creation & Sharing Potential (Low -> High)
y = [3, 6, 2, 4, 8, 5, 4, 5, 9]

# Plot each point and add labels
ax.scatter(x[:-1], y[:-1], color='b', label='Competitors')  # Competitors in blue

# Highlight "우리의 앱" with a different color and size
ax.scatter(x[-1], y[-1], color='r', s=100, label='우리의 앱', edgecolor='black', linewidth=2)

# Annotating points with competitor names
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Adding titles and labels
ax.set_title('Competitor Comparison: Technology vs Content Creation')
ax.set_xlabel('Technology Integration (Low -> High)')
ax.set_ylabel('Content Creation & Sharing Potential (Low -> High)')

# Adding a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Competitors and our app labels
labels = ['StoryCorps', 'Healium', '아름다운 이야기 할머니', 'Memoir Writing Services', 
          'AI Story Writer', '오이', '시놀', '포페런츠', '우리의 앱']

# First subplot: Technology Integration vs Content Creation & Sharing Potential
x_tech = [2, 9, 1, 3, 7, 4, 5, 3, 8]  # Technology Integration (Low -> High)
y_content = [3, 6, 2, 4, 8, 5, 4, 5, 9]  # Content Creation & Sharing Potential (Low -> High)

# Plot for competitors in first subplot
ax1.scatter(x_tech[:-1], y_content[:-1], color='gray', s=50, label='Competitors', alpha=0.6)  
# Highlight our app
ax1.scatter(x_tech[-1], y_content[-1], color='r', s=200, label='우리의 앱', edgecolor='black', linewidth=3, marker='*')

# Annotating the points with competitor names
for i, label in enumerate(labels):
    ax1.annotate(label, (x_tech[i], y_content[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Titles and labels for the first subplot
ax1.set_title('기술력 vs 콘텐츠 제작 및 공유 가능성', fontsize=14)
ax1.set_xlabel('기술력 (낮음 -> 높음)', fontsize=12)
ax1.set_ylabel('콘텐츠 제작 및 공유 가능성 (낮음 -> 높음)', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend()

# Second subplot: Senior Problem Solving Potential vs Cost
x_senior = [5, 4, 7, 3, 8, 5, 6, 4, 9]  # Senior Problem Solving Potential (Low -> High)
y_cost = [7, 8, 5, 9, 6, 3, 4, 4, 5]  # Cost (Low -> High)

# Plot for competitors in second subplot
ax2.scatter(x_senior[:-1], y_cost[:-1], color='gray', s=50, label='Competitors', alpha=0.6)
# Highlight our app
ax2.scatter(x_senior[-1], y_cost[-1], color='r', s=200, label='우리의 앱', edgecolor='black', linewidth=3, marker='*')

# Annotating the points with competitor names
for i, label in enumerate(labels):
    ax2.annotate(label, (x_senior[i], y_cost[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Titles and labels for the second subplot
ax2.set_title('노인문제 해결 가능성 vs 비용', fontsize=14)
ax2.set_xlabel('노인문제 해결 가능성 (낮음 -> 높음)', fontsize=12)
ax2.set_ylabel('비용 (낮음 -> 높음)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
